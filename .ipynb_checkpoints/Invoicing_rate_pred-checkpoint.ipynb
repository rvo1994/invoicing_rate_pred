## Cleaning

A bit of cleaning is required in order to visualize data.

```python
#Clean data and return 2 df: 1 for 2020 expenses and 1 for expenses of previous years
def clean_data(data):
    
    df = data.copy()
    
    #set column names to lowercase and remove quotes
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.strip("'")
    
    #set index to expense code
    df.set_index('code_dep', inplace=True)
    
    #change type from object to datetime
    df['date_souh_e'] = pd.to_datetime(df['date_souh_e'], format='%Y-%m-%d')
    df['date_export'] = pd.to_datetime(df['date_export'], format='%Y-%m-%d')
    
    #change type from object to float
    float_columns = ['e_originel',
                 'mhtva',
                 'encours',
                 'engage',
                 'demande_e',
                 'liquide',
                 'demande_l_prio_n',
                 'solde_l_planif',
                 'e_dem_a_date',
                 'l_prev_a_date',
                 'delta_e_a_date',
                 't1_tech',
                 't2_tech',
                 't3_tech',
                 'oct_tech',
                 'nov_tech',
                 'dec_tech',
                 'nplus1_tech',
                 'nplus2_tech',
                 'nplus3_tech',
                 'nplus4_tech',
                 'nplus5_tech']
    for column in float_columns:
        df[column] = pd.to_numeric(df[column])
    
    #remove expenses without a year
    df = df[~df['annee'].isna()]
    
    #remove forecast expenses
    current_year = datetime.datetime.now().year
    df_past = df[df['annee'] < current_year]
    
    #remove overbooking expenses
    df_cleaned = df_past[df_past['priorite'] == 'Prioritaire']
    
    #remove expenses with no commitment
    df_cleaned = df_cleaned[~((df_cleaned.encours == 0) & (df_cleaned.demande_e == 0))]
    
    #extract month from date_souh_e
    df_cleaned['month'] = df_cleaned['date_souh_e'].dt.month
    df_cleaned = df_cleaned[~df_cleaned['month'].isnull()]
    
    #create target: % of purchase order invoiced in the current year
    df_cleaned.loc[:,'target'] = df_cleaned.loc[:,'liquide'] / (df_cleaned.loc[:,'demande_e'] + df_cleaned.loc[:,'encours'])
    
    #remove expenses where the target is negative or greater than 1
    df_cleaned = df_cleaned[(df_cleaned['target'] <= 1) & (df_cleaned['target'] >= 0)]
    
    #create new DataFrame with 2020 expenses
    df_ne = df_cleaned[df_cleaned['annee'] == 2020]
    
    #create new DataFrame with expenses from previous years
    df_encours = df_cleaned[df_cleaned['annee'] < 2020]
    
    return df_ne, df_encours
df_ne, df_encours = clean_data(data)
```

```python
#define X (= features / independent variables) and y (= target / dependent variable)
X = df_ne.drop(columns=['target'])
y = df_ne['target']
```

```python
#preprocess features through a pipeline 
num_transformer = Pipeline([('robustscaler', RobustScaler())])

cat_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(sparse=False, handle_unknown = 'ignore'))])

preprocessor = ColumnTransformer([
    ('cat_transformer', cat_transformer, ['type_dep', 
                                        'nature', 
                                        'type_activite',
                                        'categorie',
                                        'versions',
                                        'code_activite',
                                        'ua_code',
                                        'tiers',
                                        'code_gest'
                                         ]),
    ('num_transformer', num_transformer, ['demande_e']
    )],
    remainder='passthrough')
```

```python
#splitting the dataset into a train (70%) and a test (30%) set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#preprocessing the train set
X_preprocessed = preprocessor.fit_transform(X_train)
```

## Model selection and training

In this section I will try different models to see which gives the best results. The aim is to get a model that is able to generalize well. Therefore we will perform cross validation on the trainset to avoid overfitting. 

The metric I am considering is negative mean absolute error. The greater this metric (the closer to zero) the better. As the target is a percentage, the mean absolute error can be interpreted as a percentage too.

### ElasticNet

ElasticNet is a linear regression model with combined L1 and L2 priors as regularizer.

```python
#perform CV and show negative MAE on validation set
en = ElasticNet(alpha=0.0001, max_iter=2000, l1_ratio=0.5)
results = cross_validate(en, X_preprocessed, y_train, cv=10, scoring= 'neg_mean_absolute_error')
np.mean(results['test_score'])
```

### KNNRegressor

KNNRegressor is a regression model based on k-nearest neighbors.

```python
#perform CV and show negative MAE on validation set
neigh = KNeighborsRegressor(n_neighbors=3)
results = cross_validate(neigh, X_preprocessed, y_train, cv=10, scoring= 'neg_mean_absolute_error')
np.mean(results['test_score'])
```

### RandomForestRegressor

A random forest is a meta estimator that fits a number of classifying decision trees on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting. 

```python
#define parameters to search (I started broad and narrowed it down to the parameters below)
rf = RandomForestRegressor()
#number of trees in random forest
n_estimators = [200,500]
#number of features to consider at every split
max_features = ['sqrt','auto']
#maximum number of levels in tree
max_depth = [30,50,None]
#create the random grid
random_grid = {'n_estimators': n_estimators,
                'max_features': max_features,
                'max_depth': max_depth}
```

```python
#random search of parameters, using 3 fold cross validation, 
#search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, 
                               param_distributions = random_grid, 
                               n_iter = 100,
                               cv = 10, 
                               scoring = 'neg_mean_absolute_error',
                               verbose=1, 
                               random_state=42,
                              n_jobs = -1)
#fit the random search model
rf_random.fit(X_preprocessed, y_train)
```

```python
#show the best parameters
rf_random.best_params_
```

```python
#show the negative MAE with the best_score
rf_random.best_score_
```

### MLPRegressor

A MLP (multi layer perceptron) is a neural network with dense layers which uses backpropagation.

```python
#MLP with 5 layers (first layer has 100 neurons and the 4 others have 10)
regr = MLPRegressor(hidden_layer_sizes=(100,10,10,10,10), batch_size=10, alpha=0.005)
results = cross_validate(regr, X_preprocessed, y_train, cv=5, scoring= 'neg_mean_absolute_error')
np.mean(results['test_score'])
```

### Xgboost

Xgboost provides a parallel tree boosting algorithm.

```python
#define parameters to search (I started broad and narrowed it down to the parameters below)
xgb1 = XGBRegressor()
parameters = {'learning_rate': [.05,0.07], #so called `eta` value
              'max_depth': [30],
              'subsample': [0.2, 0.5, 0.8],
              'colsample_bytree': [0.9],
              'n_estimators': [200]}

xgb_grid = GridSearchCV(xgb1,
                        parameters,
                        cv = 10,
                        n_jobs = -1,
                        verbose=1,
                        scoring = 'neg_mean_absolute_error',)

xgb_grid.fit(X_preprocessed,
         y_train)
```

```python
#let's have a look at the learning curves
#We can see that the model is starting to converge but it's also overfitting
train_sizes, train_scores, validation_scores = learning_curve(estimator = xgb_grid.best_estimator_,
                                                              X = X_preprocessed,
                                                              y = y_train,
                                                              cv = 10,
                                                              scoring='neg_mean_absolute_error',
                                                              n_jobs=-1)
train_scores_mean = np.mean(train_scores, axis=1)
validation_scores_mean = np.mean(validation_scores, axis=1)
plt.plot(train_sizes, train_scores_mean, label = 'Training score')
plt.plot(train_sizes, validation_scores_mean, label = 'Validation score')
plt.ylabel('neg_mean_absolute_error', fontsize = 14)
plt.xlabel('Training set size', fontsize = 14)
plt.title('Learning curves', fontsize = 18, y = 1.03)
plt.legend()
```

## Conclusion

In this notebook I preprocessed that data and tried 5 different models. The best model (Xgboost) only achieves a MAE of 20% on the test set. For this model, the learning curves showed that the model is overfitting, even though it is slowly converging. 
I tried different solutions to tackle overfitting:
* Feature selection
* Parameter tuning
* regularization of loss function

However, these solutions had little impact on the metric. In the end, considering the diversity of the expenses and the learning curves, the train set seems to be way to small (only 1627 observations) to catch the trends in the underlying expense "population". Instead the model is capturing the noise in the train set which leads to overfitting.

Because the financial tool is only in production since 2020, I could not gather additional data. At the end of 2021, the amount of data available should double and this should significantly improve the model's performance. Thanks to the pipeline, this tasks should be very easy. 

Furthermore, as I have found out during the data visualization part, the target (=invoicing rate) has a trimodal distribution (a lot of expenses are fully invoiced in the same year they are committed or not at all and a small portion of expenses are billed at around 50%). It would be worth to try a model more suited for multimodal distributed targets, such as a Mixture Density Network (MDN). This model will predict the mean, standard deviation and mixing coefficients for every observations and then based on these statistics, make a prediction for the target (see diagram hereunder).
